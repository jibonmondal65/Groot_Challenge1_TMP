Merging dataframes

```python
#deleting duplicates customer id :
rd1_unique = rd1.drop_duplicates(subset='Customer_ID')
rd2_unique = rd2.drop_duplicates(subset='Customer_ID')
rd3_unique= rd3.drop_duplicates(subset="Customer_ID")

#merging two dataframes for getting the common customers among new and regular one.So we can say they turn into regular from new later
rd_merge_2023_new_regular = pd.merge(rd1_unique, rd2_unique, on="Customer_ID", how='inner')
rd_merge_2023_new_premium = pd.merge(rd1_unique, rd3_unique, on="Customer_ID", how='inner')
rd_merge_2023_regular_premium = pd.merge(rd2_unique, rd3_unique, on="Customer_ID", how='inner')
```
